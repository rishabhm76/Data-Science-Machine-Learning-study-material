## Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

## Read Data

```python
data = pd.read_csv('1.01. Simple linear regression.csv')
data
```

## Declare Independent and Dependent Variables

```python
x = data['SAT']
y = data['GPA']
```

```python
x.shape
```

#### X is a vector so we are changing it to matrix, this is only for this simple linear regression model where we have only one feature, as sklearn assumes we would be having multiple features

```python
x_matrix = x.values.reshape(-1,1)
x_matrix.shape
```

```python
y.shape
```

## Regression 

```python
reg = LinearRegression()
```

```python
reg.fit(x_matrix,y)
```

## R squared

```python
reg.score(x_matrix,y)
```

## Coefficients

```python
reg.coef_
```

## Intercept

```python
reg.intercept_
```

## Making Predictions

```python
reg.predict([[1740]])
```

## Creating sample data and Predicting value

```python
new_data = pd.DataFrame(data=[1740,1820], columns=['SAT'])
new_data
```

```python
reg.predict(new_data)
```

```python
new_data['Predicted Value'] = reg.predict(new_data)
new_data
```

## Plotting Data

```python
plt.scatter(x,y)
yhat = reg.coef_*x_matrix + reg.intercept_
# yhat is the euation
fig = plt.plot(x, yhat, lw=4, c = 'orange', label = 'Regression Line')
plt.xlabel('SAT', fontSize = 20)
plt.ylabel('GPA', fontSize = 20)
plt.show()
```
